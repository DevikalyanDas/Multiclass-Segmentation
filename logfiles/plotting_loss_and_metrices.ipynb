Loss Plots

```
# Reading the evaluation metrics from csv and plotting
import pandas as pd
import matplotlib.pyplot as plt

# Reading CSV metrics file for task 3
loss_path_3 = '/content/loss_3.csv' # Path where the evaluation metrics is saved
loss_path_2 = '/content/loss_2.csv' # Path where the evaluation metrics is saved
loss_path_1 = '/content/loss_1.csv' # Path where the evaluation metrics is saved
fields_loss = ['epoch','train_loss','val_loss']
loss_df_3 =  pd.read_csv(loss_path_3,sep=',',usecols=fields_loss)
loss_df_2 =  pd.read_csv(loss_path_2,sep=',',usecols=fields_loss)
loss_df_1 =  pd.read_csv(loss_path_1,sep=',',usecols=fields_loss)
# Plotting the evaluation metrics
fig, ax1 = plt.subplots(1,2, figsize=(10, 4))  # defining axis and subplots

ax1[0].plot( loss_df_1['epoch'],loss_df_1['train_loss'])
ax1[0].plot( loss_df_2['epoch'],loss_df_2['train_loss'])
ax1[0].plot( loss_df_3['epoch'],loss_df_3['train_loss'])

ax1[1].plot( loss_df_1['epoch'],loss_df_1['val_loss'])
ax1[1].plot( loss_df_2['epoch'],loss_df_2['val_loss'])
ax1[1].plot( loss_df_3['epoch'],loss_df_3['val_loss'])

# ax1[0].legend(['Task_2','Task_3'],title='Models',loc=(1.1,0.5),fontsize=12,ncol=1) # add legends to the plot for different classes
ax1[0].set_title('Plot of Train Loss',loc='center',fontsize=15) # give a title
ax1[0].set_xlabel('Epoch')
ax1[0].set_ylabel('Loss Value')

ax1[1].set_title('Plot of Val Loss',loc='center',fontsize=15) # give a title
ax1[1].set_xlabel('Epoch')
ax1[1].set_ylabel('Loss Value')

fig.legend(['U-Net','R2U-Net','DDU-Net'],title="Models", loc = (0.86,0.65), ncol=1 )
fig.legend(['U-Net','R2U-Net','DDU-Net'],title="Models", loc = (0.36,0.65), ncol=1 )
fig.tight_layout()
plt.show() 
# fig.savefig('/content/loss_plots.png',bbox_inches='tight',dpi = 100)
```

```
# Reading the evaluation metrics from csv and plotting
import pandas as pd
import matplotlib.pyplot as plt

# Reading CSV metrics file for task 3
metrice_train_path_3 = '/content/train_metrices_3.csv' # Path where the evaluation metrics is saved
metrice_train_path_2 = '/content/train_metrices_2.csv' # Path where the evaluation metrics is saved
metrice_train_path_1 = '/content/train_metrices_1.csv' # Path where the evaluation metrics is saved

metrice_val_path_3 = '/content/val_metrices_3.csv' # Path where the evaluation metrics is saved
metrice_val_path_2 = '/content/val_metrices_2.csv' # Path where the evaluation metrics is saved
metrice_val_path_1 = '/content/val_metrices_1.csv' # Path where the evaluation metrics is saved

fields_metrice = ['epoch','acc','iou','dice','sens','spec']

metrice_df_train_3 =  pd.read_csv(metrice_train_path_3,sep=',',usecols=fields_metrice)
metrice_df_train_2 =  pd.read_csv(metrice_train_path_2,sep=',',usecols=fields_metrice)
metrice_df_train_1 =  pd.read_csv(metrice_train_path_1,sep=',',usecols=fields_metrice) 

metrice_df_val_3 =  pd.read_csv(metrice_val_path_3,sep=',',usecols=fields_metrice)
metrice_df_val_2 =  pd.read_csv(metrice_val_path_2,sep=',',usecols=fields_metrice) 
metrice_df_val_1 =  pd.read_csv(metrice_val_path_1,sep=',',usecols=fields_metrice) 
```
